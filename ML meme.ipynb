```python
# Import the necessary libraries
from PIL import Image

# Load the image (make sure to provide the correct path to your image file)
img = Image.open('ML.jpg')

# Display image size and mode
print("Image Size:", img.size)
print("Image Mode:", img.mode)

# Display the image
img.show()  # This will open the image in your default image viewer

# To display inside the notebook
from IPython.display import display
display(img)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image
img = mpimg.imread('ML.jpg')

# Display the image
plt.imshow(img)
plt.axis('on')  # Hide axes

# Save the image to the current directory or a specific path
plt.savefig('ML meme.jpg', bbox_inches='tight')
```
